```python
#hàm tổng hợp
df_T5T6.aggregate({'T5': ['min', 'max'], 'T6' : ['sum', 'mean']})
```

```python
#groupby
df_T5T6.groupby(['KV'])['KV'].agg(['count'])
```

```python
df_T5T6.groupby(['KV'])['T5'].agg(['min', 'max', 'sum', 'mean'])
```

```python
df_T5T6.groupby(['KV'])['T5', 'T6'].agg(['min', 'max'])
```

```python
df_T5T6.groupby(['KV', 'KT'])['T5', 'T6'].agg(['min', 'max'])
```

```python
# Pivot
pd.pivot_table(df_T5T6, values=['T5','T6'], columns='KV', aggfunc=['min', 'max', 'sum', 'mean'])
```

```python
pd.pivot_table(df_T5T6, values=['T5','T6'], columns=['KV', 'KT'], aggfunc=['min', 'max', 'sum', 'mean'])
```
